### Explore the labeled Sentinel-2 dataset

```python
# Load modules 
import pandas as pd
import os
from matplotlib import pyplot as plt
import seaborn as sns

# Reset working directory to top folder
os.chdir("..")
os.chdir("..")
print(os.getcwd())
```

```python
# Open the labeled dataset
label_df = pd.read_csv(r"data\processed\labeled_df.csv")

# Convert to long format
label_df_long = pd.melt(label_df, id_vars="label", var_name='band')
```

```python
# Number of rows for each label 
label_df.groupby("label").size().reset_index()
```

```python
# Plot the distribution of values for each band 
label_df.drop(columns='index').hist(figsize=[10,10])
plt.show()
```

```python
# Plot a correlation matrix
sub = label_df.drop('label', axis=1)
corr_matrix = sub.corr()

# Create a heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()
```
